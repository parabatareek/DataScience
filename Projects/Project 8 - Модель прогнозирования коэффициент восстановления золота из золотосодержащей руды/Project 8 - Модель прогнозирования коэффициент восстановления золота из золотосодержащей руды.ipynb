# Проверка расчета эффективности обогащения Au на этапе флотации #

<a id='func_flotation_rate'><a/>
## Функция расчета эффективности обогащения ##

```python
def recovery_rate(raw):
    return (raw['output']*(raw['input'] - raw['tail']))/(raw['input']*(raw['output'] - raw['tail']))*100
```

<a id='func_flotation_rate_prepare'></a>
## Функция подготовки данных для расчета ##

```python
def recovery_data_return(input, output, tail):
    return pd.DataFrame({'input' : input,
                         'output' : output,
                         'tail' : tail}).reset_index(drop=True)
```

<a id='calc_flotation_MAE'></a>
## Вычисление MAE для этапа флотации ##

```python
rougher_recovery_data = recovery_data_return(input=df_gold_train['rougher.input.feed_au'],
                                             output=df_gold_train['rougher.output.concentrate_au'],
                                             tail=df_gold_train['rougher.output.tail_au'])

rougher_recovery_data['rougher.output.recovery'] = rougher_recovery_data.apply(recovery_rate, axis=1)

print(f"""MAE для процедуры флотации Au составляет: {mean_absolute_error(df_gold_train['rougher.output.recovery'],
                                                                         rougher_recovery_data['rougher.output.recovery'])}""")
```

## Вывод ##
Малая величина метрики MAE говорит о том что расчитанное и реальное значение эффективности обогащения отличаются очень слабо.

<a id='model_func_feature_scale'></a>
## Функция масштабирования признаков ##

```python
def func_features_scaler(features_train, features_test):
    scaler = StandardScaler()

    scaler.fit(features_train)
    features_train = scaler.transform(features_train)
    features_test = scaler.transform(features_test)

    return features_train, features_test
```

<a id='model_func_hyperparams'></a>
## Функция получения набора гиперпараметров ##

```python
def func_get_search_params(clf_name):
    params = {'DecisionTreeRegressor' : {'criterion' : ['mse', 'friedman_mse', 'mae'],
                                         'splitter'  : ['best', 'random'],
                                         'max_depth' : range(1, 21, 4),
                                         'min_samples_split' : range(2,21,4),
                                         'min_samples_leaf' : range(1,21,4),
                                         'max_features' : ['auto']},
              'RandomForestRegressor' : {'n_estimators': range(1,21,4),
                                         'criterion' : ['mse', 'mae'],
                                         'max_depth' : range(1, 21, 4),
                                         'min_samples_split' : range(2,21,4),
                                         'min_samples_leaf' : range(1,21,4),
                                         'max_features' : ['auto']}}
    return params[clf_name]
```

<a id='model_func_smape'></a>
## Функция расчета sMAPE ##

```python
def func_smape(target_real, target_predict):
    smape = sum(abs(target_real - target_predict)/((abs(target_real) + abs(target_predict))/2))*100/len(target_real)  
    return smape
```

## Проверка модели "Случайный лес" на тестовом наборе данных ##

```python
# Вычисление SMAPE для этапа флотации ##
clf = RandomForestRegressor(n_estimators=rougher_random_forset_params.get('n_estimators'),
                            criterion=rougher_random_forset_params.get('criterion'),
                            max_depth=rougher_random_forset_params.get('max_depth'),
                            min_samples_split=rougher_random_forset_params.get('min_samples_split'),
                            min_samples_leaf=rougher_random_forset_params.get('min_samples_leaf'),
                            max_features=rougher_random_forset_params.get('max_features'),
                            random_state=state)
                            
clf.fit(X=rougher_features_train_scale,
        y=rougher_target_train)

rougher_prediction_test = pd.Series(clf.predict(rougher_features_test_scale))
smape_rougher = func_smape(target_real=rougher_target_test, target_predict=rougher_prediction_test)

# Вычисление SMAPE для этапа финальной очистки ##
clf = RandomForestRegressor(n_estimators=final_random_forset_params.get('n_estimators'),
                            criterion=final_random_forset_params.get('criterion'),
                            max_depth=final_random_forset_params.get('max_depth'),
                            min_samples_split=final_random_forset_params.get('min_samples_split'),
                            min_samples_leaf=final_random_forset_params.get('min_samples_leaf'),
                            max_features=final_random_forset_params.get('max_features'),
                            random_state=state)

clf.fit(X=final_features_train_scale, 
        y=final_target_train)

final_prediction_test = pd.Series(clf.predict(final_features_test_scale))
smape_final = func_smape(target_real=final_target_test, target_predict=final_prediction_test)

final_smape = 0.25*smape_rougher + 0.75*smape_final
print(f'Итоговое значение sMAPE: {final_smape}')
```
